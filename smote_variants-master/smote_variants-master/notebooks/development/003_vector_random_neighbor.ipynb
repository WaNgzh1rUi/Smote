```python
import numpy as np
```

```python
dists = np.random.rand(100, 5)
dists = (dists.T/np.sum(dists, axis=1)).T
```

```python
dists
```

```python
def random_neighbors(dists):
    prob_cum = np.cumsum(dists, axis=1)
    r = np.random.rand(dists.shape[0])
    return np.argmax((r < prob_cum.T).T, axis=1)
```

```python
%%timeit
random_neighbors(dists)
```

```python
indices = list(range(5))
```

```python
%%timeit
np.apply_along_axis(lambda x: np.random.choice(indices, p=x), axis=1, arr=dists)
```

```python
results = []
for _ in range(10000):
    results.append(random_neighbors(dists)[0])
```

```python
indices, counts = np.unique(results, return_counts=True)
```

```python
def cartesian_product(*arrays):
    la = len(arrays)
    dtype = np.result_type(*arrays)
    arr = np.empty([len(a) for a in arrays] + [la], dtype=dtype)
    for i, a in enumerate(np.ix_(*arrays)):
        arr[...,i] = a
    return arr.reshape(-1, la)

weights = np.array([[1, 2, 3],
                    [4, 5, 6]]).T
weights = weights / np.sum(weights)

coords = cartesian_product(np.array([0, 1, 2]), np.array([0, 1]))
coords[np.random.choice(np.arange(len(coords)), 20, p=weights[coords[:,0], coords[:,1]])]
```
